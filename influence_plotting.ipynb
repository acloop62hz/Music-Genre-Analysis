```python
#radar plotting
import csv
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from scipy import stats
import math
import matplotlib.pyplot as plt

dict={}
f1 = open('data/artist_music_genre_data.csv','r',encoding="UTF-8")
reader = csv.reader(f1)

flag=0
genre='#'
for line in reader:
    if line[0]== 'artist_id':continue
    if line[17] == '':break
    if line[15] != genre:
        genre=line [15]
        dict.update({genre:[]})
    else:
        dict[genre].append(int(line[16]))

f1.close()

df=pd.read_csv('data/artist_music_genre_data.csv')
features=['danceability','energy','valence','tempo','loudness','mode','key','acousticness','instrumentalness','liveness','speechiness']
for i in range(0,len(features)):
    s = np.ones_like(df[features[i]])
    s = s[:, None]
    for k in range(0, len(df[features[i]])):
        if df[features[i]][k] < 0:
            temp=df[features[i]][k]+41
        else:temp=df[features[i]][k]
        s[k] = np.sqrt(temp)
    scaler = preprocessing.StandardScaler()
    std=scaler.fit_transform(s)
    minmax_scale = preprocessing.MinMaxScaler(feature_range=(0, 1))
    sc=minmax_scale.fit_transform(std)
    if(flag == 0):
        mtx=sc
        flag = 1
    else: mtx=np.c_[mtx,sc]

s = np.array(df['count'])
s = s[:, None]
mtxc=np.c_[mtx,s]
data=pd.DataFrame(mtxc)

dict_mean={}
for header in dict.keys():
    dict_mean.update({header:[]})
    for k in range(0,10):
        sum=0
        count=0
        for i in range(0,len(dict[header])):
            sum=sum+data[k][dict[header][i]-1]*data[11][dict[header][i]-1]
            count=count+data[11][dict[header][i]-1]
        dict_mean[header].append(sum/count)



dict_std=[]
for i in range(0,11):
    newarry = []
    for header in dict.keys():
        newarry.append(dict_mean[header][i])
    newarry=np.array(newarry)
    stdd=np.std(newarry)
    dict_std.append(stdd)




name='Avant-Garde'
plt.style.use('ggplot')
angles=np.linspace(0, 2*np.pi,len(dict_mean[name]), endpoint=False)
values=np.concatenate((dict_mean[name],[dict_mean[name][0]]))
angles=np.concatenate((angles,[angles[0]]))
fig=plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, values, 'o-', linewidth=2)
# 填充颜色
ax.fill(angles, values, alpha=0.25)

# 设置图标上的角度划分刻度，为每个数据点处添加标签
ax.set_thetagrids(angles * 180 / np.pi, features)

# 设置雷达图的范围
ax.set_ylim(0, 1.1)
# 添加标题
plt.title(name)
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#data preprocessing
data = pd.read_table("data/sort_genre_year.csv", sep=",").sort_values(by=["artist_active_start", "artist_main_genre", "artist_id"])
genre_table = data.groupby(["artist_main_genre", "artist_active_start"]).agg({"artist_id":"count"}).unstack(fill_value=0).rename_axis(None)["artist_id"]
genre_table = genre_table.drop(["Unknown"])
genre_table.to_csv('./count_genre_year.csv', sep=',', header=True, index=True)
```

```python
artists = pd.read_table("data/artists.csv", sep=",").set_index(['artist_id']).sort_values(by=["influence"], ascending=False)
#rock
artists.loc[(artists["artist_main_genre"] == "Pop/Rock") 
            & ((artists["artist_active_start"] == 1950)
              | (artists["artist_active_start"] == 1960))].head()
```

```python
#Electronic
artists.loc[(artists["artist_main_genre"] == "Electronic") 
            & ((artists["artist_active_start"] == 1970)
              | (artists["artist_active_start"] == 1960))].head()
```

```python
#R&B
artists.loc[(artists["artist_main_genre"] == "R&B;") 
            & ((artists["artist_active_start"] == 1940)
              | (artists["artist_active_start"] == 1950))].head()
```

```python
#decay genre
dfArt = artists.loc[(artists["artist_main_genre"] == "Jazz") 
            | (artists["artist_main_genre"] == "Vocal")
            | (artists["artist_main_genre"] == "Blues")].head(10)
dfArt
```

```python
infl = pd.read_table("data/influence_data.csv", sep=",").set_index(['influencer_id'])
infl.loc[dfArt.index.tolist()]
countFollow = infl.loc[dfArt.index.tolist()].groupby(["follower_main_genre"]).agg({"follower_id": "count"}).sort_values(by=["follower_id"], ascending=False)
countFollow.columns = ["count"]
countFollow.drop(["Jazz", "Vocal", "Blues"])
```
